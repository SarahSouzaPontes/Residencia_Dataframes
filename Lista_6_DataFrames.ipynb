````python
base1 = pd.read_json("https://storage.googleapis.com/pedagogico/formacao-dados/tabela_vendas.json")
print(base1)
````

````python
#criando coluna com o total
base1['total'] = base1['quantidade'].mul(base1['valor_produto'])
````

````python
#simulando cenários
#Aumento de 5%
#Criar total + 5%
base1['total_mais_5p'] = base1['total'].mul(1.05)
````

````python
#Criar total - 10%
base1['total_menos_10p'] = base1['total'].mul(0.9)
````

````python
#media
base1[['total_mais_5p']].mean()
#mediana
base1[['total_mais_5p']].median()
````

````python
#media
base1[['total_menos_10p']].mean()
#mediana
base1[['total_menos_10p']].median()
````
